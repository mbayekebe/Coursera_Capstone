# Scraping Wiki web page

```python
# import the library used to query site
import requests

# specify the url
wikiurl = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

#query the website and return the html to the variable wikipage
wikipage = requests.get(wikiurl)
```

```python
# import Beautiful soup functions to parse the data returned from the site
from bs4 import BeautifulSoup

# parse the html in the wikipage variable, store it in BeautifulSoup format
soup = BeautifulSoup(wikipage.content, 'html.parser')
```

```python
# Find the table
table = soup.find_all('table', class_='wikitable sortable')
```

```python
# import 
import pandas as pd

df = pd.read_html(str(table), header=0)
dfs = df[0]

```

```python
dfs
```

```python
dfs.to_csv('Toronto.csv')
newdf = pd.read_csv('Toronto.csv', index_col=0)
newdf
type(newdf)
```

```python
newdf.rename(columns = {'Postcode':'PostalCode','Borough':'Borough','Neighbourhood':'Neighborhood'}, inplace=True)
```

```python
# check again iterate and look for neighborhood not assigned
for row in newdf.itertuples():
    if (row.Neighborhood == 'Not assigned') & (row.Borough != 'Not assigned'):
        print(row, row.Neighborhood, row.Borough)
        
```

```python
# assigning 
newdf.iat[8, 2] = "Queen's Park"
```

```python
# checking assignment
newdf.iat[8, 2]
```

```python
# ignore borough Not assigned
newdf.drop(newdf.loc[newdf['Borough']=='Not assigned'].index, inplace=True)

```

```python
newdf
```

```python

import numpy as np
# initialize working lists
sl_Neighborhood = []
Rec_line = []
ResArray = []

strNeighborhood = ''

# iterate through dataframe
Prev_ID = newdf.iloc[0]['PostalCode']
Prev_BR = newdf.iloc[0]['Borough']
Prev_NB = newdf.iloc[0]['Neighborhood']

strNeighborhood = Prev_NB

for i in range(1, len(newdf)):
    
    Curr_ID = newdf.iloc[i]['PostalCode']
    Curr_BR = newdf.iloc[i]['Borough']
    Curr_NB = newdf.iloc[i]['Neighborhood']
    # compare
    if Curr_ID == Prev_ID:
     
     #sl_Neighborhood.append(newdf.iloc[i]['Neighborhood'])
     strNeighborhood = strNeighborhood+','+ newdf.iloc[i]['Neighborhood']
    
    else:
      
      sl_Neighborhood.append(Prev_NB)
      Rec_line.append(Prev_ID)
      Rec_line.append(Prev_BR)
      Rec_line.append(strNeighborhood)  
      #Rec_line.append(sl_Neighborhood)
      #print(Rec_line)
      ResArray.append(Rec_line)
        
      Prev_ID = Curr_ID
      Prev_BR = Curr_BR
      Prev_NB = Curr_NB
      strNeighborhood = Curr_NB
      sl_Neighborhood = []
      Rec_line = []
      

Finaldata = pd.DataFrame(ResArray)    
Finaldata.columns=['PostalCode','Borough','Neighborhood']
Finaldata
    
```
